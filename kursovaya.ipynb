```python
import csv 
from collections import defaultdict 


# Определяем путь к CSV файлу 
filecsv = 'magazin.csv' 

# Читаем данные из CSV файла 
def read_sales_data(filepath): 
    with open('magazin.csv','r', encoding='utf-8') as file: 
        csv_reader = csv.DictReader(file) 
        sales_data = [] 
        for row in csv_reader: 
            sales_data.append(row) 
    return sales_data 
 
# Выводим данные на экран 
def print_sales_data(sales_data): 
    for sale in sales_data: 
        print(f"Номер заказа: {sale['Номер заказа']}") 
        print(f"Дата заказа: {sale['Дата заказа']}") 
        print(f"Название товара: {sale['Название товара']}") 
        print(f"Категория товара: {sale['Категория товара']}") 
        print(f"Количество продаж: {sale['Количество продаж']}") 
        print(f"Цена за единицу: {sale['Цена за единицу']}") 
        print(f"Общая стоимость: {sale['Общая стоимость']}") 
        print("-" * 40) 
 
if __name__ == "__main__": 
    sales_data = read_sales_data(filecsv) 
    print_sales_data(sales_data)
```

```python
#Общая выручка магазина 
def calculate_total_revenue(filecsv): 
    total_revenue = 0 
     
    with open('magazin.csv','r', encoding='utf-8') as file: 
        reader = csv.DictReader(file) 
        for row in reader: 
            total_revenue += int(row['Общая стоимость']) 
     
    return total_revenue 
 
filecsv = 'magazin.csv' 
total_revenue = calculate_total_revenue(filecsv) 
print(f'Общая выручка: {total_revenue} рублей')
```

```python
#Товар который принес наибольшую выручку 
def calculate_total_revenue(csv_file): 
    revenue_by_item = {}  # Хэш-таблица для отслеживания общей выручки от каждого товара 
 
    # Читаем данные из файла CSV и подсчитываем общую выручку от каждого товара 
    with open(csv_file, 'r', newline='', encoding='utf-8') as file: 
        csv_reader = csv.DictReader(file) 
        for row in csv_reader: 
            item = row['Название товара'] 
            revenue = int(row['Общая стоимость']) 
            # Если товар уже присутствует в хэш-таблице, увеличиваем его выручку 
            if item in revenue_by_item: 
                revenue_by_item[item] += revenue 
            # Если товар встречается впервые, добавляем его в хэш-таблицу 
            else: 
                revenue_by_item[item] = revenue 
     
    # Находим товар с максимальной выручкой 
    max_revenue_item = max(revenue_by_item, key=revenue_by_item.get) 
    max_revenue = revenue_by_item[max_revenue_item] 
     
    return max_revenue_item, max_revenue 
 
# Путь к файлу CSV с данными о продажах 
filecsv = 'magazin.csv' 
 
# Находим товар, который принес наибольшую выручку, используя данные из файла CSV 
max_revenue_item, max_revenue = calculate_total_revenue(filecsv) 
print(f'Товар, который принес наибольшую выручку: {max_revenue_item}, выручка: {max_revenue} рублей')
```

```python
def calculate_total_sold_units(filecsv): 
    # Создаем массив для хранения количества проданных единиц каждого товара 
    total_sold_units = [] 
     
    with open(filecsv, 'r', encoding='utf-8') as file: 
        next(file)  # Пропускаем заголовок файла 
        for line in file: 
            row = line.strip().split(',') 
            sold_units = int(row[4])  # Количество проданных единиц 
            total_sold_units.append(sold_units) 
     
    # Суммируем значения в массиве, чтобы получить общее количество проданных единиц 
    total = sum(total_sold_units) 
     
    return total 
 
# Пример использования функции с файлом 
filecsv = 'magazin.csv' 
total_sold_units = calculate_total_sold_units(filecsv) 
print("Общее количество проданных единиц:", total_sold_units)
```

```python
def srednyapricezatovar(filecsv): 
    total_revenue = 0 
    total_units_sold = 0 
     
    with open(filecsv, 'r', encoding='utf-8') as file: 
        next(file)  # Пропускаем заголовок файла 
        for line in file: 
            row = line.strip().split(',') 
            x = int(row[5])  # Цена за единицу 
            y = int(row[4])  # Количество проданных единиц 
             
            # Суммируем общую выручку и общее количество проданных единиц 
            total_revenue += x * y 
            total_units_sold += y 
     
    # Вычисляем среднюю цену за единицу 
    if total_units_sold > 0: 
        average_price_per_unit = total_revenue / total_units_sold 
    else: 
        average_price_per_unit = 0 
     
    return average_price_per_unit 
 
# Пример использования функции с файлом 
filecsv = 'magazin.csv' 
average_price_per_unit = srednyapricezatovar(filecsv) 
print("Средняя цена за единицу товара:", average_price_per_unit)
```

```python
#Товар который был продан максимальное количество раз 
def bubble_sort(arr): 
    n = len(arr) 
    for i in range(n-1): 
        for j in range(0, n-i-1): 
            if arr[j][1] < arr[j+1][1]: 
                arr[j], arr[j+1] = arr[j+1], arr[j] 
    return arr  # Возвращаем отсортированный список 
 
def most_sold_item_from_file(filecsv): 
    sales_count = {}  # Создаем пустой словарь для отслеживания количества продаж каждого товара 
 
    # Читаем данные из файла CSV и подсчитываем количество продаж каждого товара 
    with open(filecsv,'r', encoding='utf-8') as file: 
        csv_reader = csv.DictReader(file) 
        for row in csv_reader: 
            item = row['Название товара'] 
            sales = int(row['Количество продаж']) 
            if item in sales_count: 
                sales_count[item] += sales 
            else: 
                sales_count[item] = sales 
     
    # Преобразуем словарь в список кортежей (название товара, количество продаж) 
    sales_list = list(sales_count.items()) 
 
    # Сортируем список по количеству продаж с помощью пузырьковой сортировки 
    sorted_sales_list = bubble_sort(sales_list) 
     
    # Выводим отсортированный список 
    print("Отсортированный список после пузырьковой сортировки:") 
    for item in sorted_sales_list: 
        print(item) 
     
    # Находим товар с максимальным количеством продаж 
    max_sold_item = sorted_sales_list[0][0] 
    max_sales = sorted_sales_list[0][1] 
     
    return max_sold_item, max_sales 
 
# Путь к файлу CSV с данными о продажах 
filecsv = 'magazin.csv' 
 
# Находим товар, который был продан наибольшее количество раз, используя данные из файла CSV 
max_sold_item, max_sales = most_sold_item_from_file(filecsv) 
print(f'Товар, который был продан наибольшее количество раз: {max_sold_item}, количество продаж: {max_sales}')
```

```python
# Максималоьная и минимальная цена товаров 
def find_min_max_price_items(filecsv): 
    items_prices = []  # Список для хранения названий товаров и их цен 
 
    # Читаем данные из файла CSV 
    with open(filecsv, 'r', encoding='utf-8') as file: 
        csv_reader = csv.DictReader(file) 
        for row in csv_reader: 
            item = row['Название товара'] 
            price = int(row['Цена за единицу']) 
            items_prices.append((item, price)) 
 
    # Сортируем список по цене товара 
    items_prices.sort(key=lambda x: x[1]) 
 
    # Находим товар с минимальной ценой 
    min_price_item = items_prices[0] 
 
    # Находим товар с максимальной ценой 
    max_price_item = items_prices[-1] 
 
    return min_price_item, max_price_item 
 
# Путь к файлу CSV с данными о продажах 
filecsv = 'magazin.csv' 
 
# Находим товар с минимальной и максимальной ценой 
min_price_item, max_price_item = find_min_max_price_items(filecsv) 
 
# Выводим найденные товары 
print(f'Товар с минимальной ценой: {min_price_item[0]}, цена: {min_price_item[1]} рублей.') 
print(f'Товар с максимальной ценой: {max_price_item[0]}, цена: {max_price_item[1]} рублей.')
```

```python
# Отчет магазина 
def generate_report(filecsv): 
    total_revenue = 0 
    total_units_sold = 0 
    sales_count = {}  # Словарь для подсчета количества проданных единиц каждого товара 
 
    # Читаем данные из файла CSV 
    with open(filecsv, 'r', encoding='utf-8') as file: 
        csv_reader = csv.DictReader(file) 
        for row in csv_reader: 
            # Подсчитываем общую выручку магазина 
            total_revenue += int(row['Общая стоимость']) 
            # Подсчитываем общее количество проданных единиц 
            total_units_sold += int(row['Количество продаж']) 
            # Подсчитываем количество проданных единиц каждого товара 
            item = row['Название товара'] 
            sales = int(row['Количество продаж']) 
            if item in sales_count: 
                sales_count[item] += sales 
            else: 
                sales_count[item] = sales 
 
    # Выводим общую выручку магазина 
    print(f'Общая выручка магазина: {total_revenue} рублей.') 
    # Выводим общее количество проданных единиц 
    print(f'Общее количество проданных единиц: {total_units_sold}.') 
 
    # Выводим информацию о количестве проданных единиц каждого товара и их доле в общей выручке 
    print('\nИнформация о продажах по каждому товару:') 
    for item, sales in sales_count.items(): 
        print(f'{item}: Продано {sales} единиц') 
 
# Путь к файлу CSV с данными о продажах 
filecsv = 'magazin.csv' 
 
# Генерируем отчет 
generate_report(filecsv)
```
